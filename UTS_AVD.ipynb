# **Analisis EDA Dataset Gaming Academic Performance**

# **Latar Belakang**

Perkembangan teknologi dan industri game membuat aktivitas bermain game menjadi bagian dari kehidupan pelajar dan mahasiswa. Meskipun game dapat menjadi sarana hiburan dan interaksi sosial, intensitas bermain yang berlebihan dapat memengaruhi performa akademik, seperti waktu belajar, kualitas tidur, tingkat stres, dan nilai akademik.

Dataset Gaming Academic Performance digunakan untuk menganalisis hubungan antara kebiasaan bermain game dengan prestasi akademik mahasiswa. Melalui proses Exploratory Data Analysis (EDA), dapat diketahui pola, distribusi data, serta hubungan antarvariabel seperti durasi bermain game, jam belajar, tingkat kecanduan, dan nilai akademik. Analisis ini penting untuk memahami dampak aktivitas gaming terhadap pencapaian akademik dan membantu menemukan keseimbangan antara hiburan dan pendidikan.

# **Tahapan Proses**

## **Import Library**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
```

## **Memuat Dataset**

```python
file = "/content/Gaming_Academic_Performance.csv"

df = pd.read_csv(file)

df
```

## **Eksplorasi Data (EDA)**

### **Comparison/Perbandingan**


Bagian ini merupakan tahap membandingkan kolom Gender dan Gaming Hours.

Tujuan: Mengetahui Gender mana yang memiliki score Gaming Hours terbanyak.

Visualisasi: Bar Chart.

```python
comparison = df.groupby('gender')['gaming_hours'].sum().sort_values(ascending=True)

plt.figure(figsize=(10,10))
sns.barplot(x=comparison.index, y=comparison.values, palette='viridis', hue=comparison.index, legend=True)
plt.title('Tingkatan Gaming Hours berdasarkan Gender')
plt.xlabel('Gender')
plt.ylabel('Gaming Hours')
plt.xticks(rotation=45)
plt.show()
```

### **Composition/Komposisi**

Aktivitas: Melihat proporsi Gaming Genre berdasarkan Sleep Hours.

Tujuan: Mengetahui genre game yang mempengaruhi waktu tidur pelajar.

Visualisasi: Pie Chart.

```python
comp = df.groupby('gaming_genre')['sleep_hours'].sum()

plt.figure(figsize=(7, 7))
plt.pie(comp.values, labels=comp.index, autopct='%1.1f%%', colors=sns.color_palette('Blues', len(comp)))
plt.title('Proporsi Gaming Genre Berdasarkan Sleep Hours')
plt.tight_layout()
plt.show()
```

### **Distribution/Distribusi**

Aktivitas: Menganalisis  frekuensi penyebaran tingkat kecanduan game dalam bentuk persentase.

Tujuan: Melihat distribusi tingkat kecanduan game seorang pelajar.

Visualisasi: Histogram.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['addiction_score'], bins=20, kde=True)
plt.title('Distribusi Kecanduan Game')
plt.xlabel('Addiction Score')
plt.ylabel('Frekuensi')
plt.show()
```

### **Relationship/Hubungan**

Aktivitas: Menganalisis korelasi data antara **(study_hours)** dan **(grades)**

Tujuan: Mengetahui apakah waktu jam belajar memengaruhi nilai seorang pelajar.
Visualisasi: Scatter Plot

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='study_hours', y='grades', alpha=0.4, color='dodgerblue')
plt.title('Hubungan Study Hours dengan Grades')
plt.xlabel('Study Hours')
plt.ylabel('Grades')
plt.tight_layout()
plt.show()
```
